```python
df = pd.read_csv('/content/drive/MyDrive/lab 4/orders.csv')
df.head(5)
```

```python
df.describe
```

```python
df.info
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y%m%d')
df.head(3)
```

```python
df['tshirt_category'].unique()
```

```python
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: str(x).replace('Wh Tshirt F','White T-Shirt F'))
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: str(x).replace('Wh Tshirt M','White T-Shirt M'))
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: str(x).replace('Bl Tshirt F','Black T-Shirt F'))
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: str(x).replace('Bl Tshirt M','Black T-Shirt M'))
df['tshirt_category'].unique()
```

# ***2 metoda***

```python
df.replace({'Wh Tshirt M':'White T-Shirt M','Bl Tshirt F':'Black T-Shirt F','Bl Tshirt M':'Black T-Shirt M', 'Wh Tshirt F':'White T-Shirt F'}, inplace=True)
df['tshirt_category'].unique()
```

```python
df[['color', 'type', 'tshirt sex']] = df.tshirt_category.str.split(" ",expand=True,)
df.head(10)
```

```python
df['color'].replace({'Hoodie': None, 'Tennis': None}, inplace=True)
df['color'].unique()
```

```python
df['type'].replace({None:'Hoodie', 'Shirt': 'Tennis Shirt'}, inplace=True)
df['type'].unique()
```

```python
df['tshirt sex'].replace({None:'Unisex', 'M': 'Male', 'F': 'Female'}, inplace=True)
df['tshirt sex'].unique()
```

```python
df.drop('tshirt_category', axis=1)
```

```python
df[df.order_date.between('31-12-2014','02-08-2016')]
```
